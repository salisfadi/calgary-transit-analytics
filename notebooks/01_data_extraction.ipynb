```python
catalog_url = "https://data.calgary.ca/resource/3fy4-eztt.json"
params = {"$q": "fleet vehicle", "$limit": 20}
response3 = requests.get(catalog_url, params=params)
results = response3.json()

for item in results:
    print(item.get("name"), "→", item.get("uid"))
```

## Data Source Decision Log

**Original plan:** Use Calgary Transit Fleet List (ekzc-4ss8) as the primary vehicle dataset.

**Issue found:** This dataset returns `403 Forbidden` with the message 
"You must be logged in to access this resource" — confirmed via direct 
browser test, not a code or authentication header issue. The dataset's 
metadata catalog also no longer indexes it under searchable terms like 
"fleet" or "vehicle," suggesting it has been deprecated or restricted 
since its original publication.

**Resolution:** Queried Calgary's open data catalog (dataset 3fy4-eztt) 
to search for currently active, structured alternatives. Two candidates 
("Annual Corporate Energy Use by Bus Units" and "Realtime Vehicle 
Positions GTFS-RT") returned valid metadata but zero columns — likely 
deprecated or served via a non-standard format (GTFS-RT uses protobuf, 
not standard JSON).

**Final dataset lineup:**
- Calgary Transit Ridership (iema-jbc4) — demand data
- Calgary Transit Routes (hpnd-riq4) — network structure
- Hours of Service Provided by Calgary Transit (64iy-rpqq) — supply/utilization data

This substitution is arguably stronger than the original plan: combining 
ridership and operating hours allows direct calculation of utilization 
(boardings per operating hour), a core metric for fleet planning.
